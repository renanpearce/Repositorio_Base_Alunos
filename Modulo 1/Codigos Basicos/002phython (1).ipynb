# Função print ()
a função  print() é uma das funções mais importantes e usadas na linguagem phython. sua função é  basicamente, exibir  mensagens na tela  ou envia-las  para outro dispositivo, como imprimir  dentro  de um arquivo de texto

no phython 3,  print () É função interna, de modo que  não é necessario importar  nenhuma  biblioteca  para poder  utiliza-la . basta chama-la e passar  os argumentos  necessarios

```python
print   ("meu primeiro print!") # todo  texto  deve estar  entre  aspas  simples  ou duplas
```

```python
 print ("meu segundo print")
```

```python
print ("d' agua '") #  quando uma pala  contem aspas  as aspas iniciais  deve ser diferentes
```

```python
print (20+15) # o phython realiza a operação  que está  demtro do parenteses
```

```python
print ("20+15") # a operação entre aspas sera apenas impressa não executada,
#pois o phython entende isso como texto
```

```python
print ("20+15",20+15)
```

```python
n1 =  "1"
n2 = "3"
print ( n1 + n2) # neste exemplo o phython faz a  concatenação (unir)  das 2 strings
```
